```python
#filtros SOBEL
sobelx = cv2.Sobel(gray_dog, cv2.CV_64F, 1, 0, ksize=5)
sobely= cv2.Sobel(gray_dog,cv2.CV_64F,0,1,ksize=5)
```

```python
#combinar las respuestas en magnitud
edges = cv2.magnitude(sobelx, sobely)
```

```python
#normalizar los valores para mostrar la imagen corretamente.
edges = cv2.normalize(edges, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
```

```python
#mostrar imagen.
cv2_imshow(edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
